```python
#test out one image from the folder

#load trained model
face_cascade = cv2.CascadeClassifier(data_file + 'simpson_cascade/cascade.xml')

img = cv2.imread(test_image_file_path + '/' + photo_list[2])
faces = face_cascade.detectMultiScale(img, 1.3, 5)
#print(img)
(x,y,w,h) = faces[0]
(x,y,w,h)
face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
plt.imshow(face_img)

#replaced and removed bad image files
```

```python
# going through each image in the test_image folder
# seeing if the face was correctly found in the image

#load trained model
face_cascade = cv2.CascadeClassifier(data_file + 'simpson_cascade/cascade.xml')

for photo in photo_list:
    print(photo)
    #read photo and set to image
    img = cv2.imread(test_image_file_path + '/' + photo)
    #rectangle cordinates of faces in image
    faces = face_cascade.detectMultiScale(img, 1.3, 5)
    #draw a rectangle around the face identified
    #detection_img = face_cascade.draw_rectangles(img, faces)
    (x,y,w,h) = faces[0]
    face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    #print(face_img)
    #plt.imshow('Unprocessed', face_img)
    plt.imshow(face_img)
    plt.show()
    
```
