```python
import os
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#Trying to see what version of cv2 is being used
print(cv2.__version__)
```

```python
# path to image test files

data_file = '/Users/lynnpowell/Documents/DS_Projects/Data_Files/'
test_image_file_path = '/Users/lynnpowell/Documents/DS_Projects/Data_Files/Simpson_Images/test_images'
```

```python
# all phots in the test_image folder

photo_list = os.listdir(test_image_file_path)
photo_list
```

```python
#removing items from list

remove_lst = ['.DS_Store', 'bart_solo_base64.txt']

for delete in remove_lst:
    photo_list.remove(delete)
    
photo_list
```

```python
#test out one image from the folder

#load trained model
face_cascade = cv2.CascadeClassifier(data_file + 'simpson_cascade/cascade.xml')

img = cv2.imread(test_image_file_path + '/' + photo_list[2])
faces = face_cascade.detectMultiScale(img, 1.3, 5)
#print(img)
(x,y,w,h) = faces[0]
(x,y,w,h)
face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
plt.imshow(face_img)

#replaced and removed bad image files
```

```python
# going through each image in the test_image folder
# seeing if the face was correctly found in the image

#load trained model
face_cascade = cv2.CascadeClassifier(data_file + 'simpson_cascade/cascade.xml')

for photo in photo_list:
    print(photo)
    #read photo and set to image
    img = cv2.imread(test_image_file_path + '/' + photo)
    #print(img[0])
    #rectangle cordinates of faces in image
    faces = face_cascade.detectMultiScale(img, 1.3, 5)
    #draw a rectangle around the face identified
    #detection_img = face_cascade.draw_rectangles(img, faces)
    if len(faces) == 0:
        print('\n')
        print('No identified image')
        print('\n')
        continue
    print(faces)
    (x,y,w,h) = faces[0]
    face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    #print(face_img)
    #plt.imshow('Unprocessed', face_img)
    plt.imshow(face_img)
    plt.show()
    
```
